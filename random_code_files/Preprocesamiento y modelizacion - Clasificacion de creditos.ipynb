## Primera Parte. Limpieza: Imputación de Missing Values


- Las columnas numéricas deben imputarse con kNN (n=5)
- Las variables categóricas deben imputarse con la moda.
- Realice el procedimiento de tal manera de **preservar los nombres de las variables**, para facilitar el siguiente paso.
- Denomine a los dataframes obtenidos como X_train_limpio, y_train, X_test_limpio, y_test


Al finalizar debería tener los DataFrames X_train_limpio y X_test_limpio, además de y_train e y_test que no se han modificado.

```python
from sklearn.model_selection import train_test_split
X=df.drop(columns=['A16'])
X_train, X_test, y_train, y_test = train_test_split(X, df['A16'], test_size=0.2, random_state=123)

#y_train = y_train.to_frame()
```

```python
from sklearn.impute import SimpleImputer  
from sklearn.compose import ColumnTransformer 
from sklearn.impute import KNNImputer

#Obtengo arrays con los tipos de columnas para faciltiar la tarea
cols_num=X_train.select_dtypes(include=['int64','float64']).columns
cols_cat=X_train.select_dtypes(include='object').columns

#Creo los imputadores
t1=('imputador_num', KNNImputer(n_neighbors=5), cols_num)  
t2=('imputador_cat', SimpleImputer(strategy='most_frequent'), cols_cat) 

#Creo el transformador
transformador_columnas = ColumnTransformer(transformers=[t1,t2],remainder='passthrough')

#Entreno el transformador
transformador_columnas.fit(X_train) 
```

```python
#TransformocX_train y X_test
X_train_limpio = transformador_columnas.transform(X_train) 

X_test_limpio = transformador_columnas.transform(X_test)

#Creo los data frames con los nombres de columnas correspondientes.
cols_orden=cols_num.tolist() + cols_cat.tolist()
X_train_limpio=pd.DataFrame(X_train_limpio, columns=cols_orden) 
X_test_limpio=pd.DataFrame(X_test_limpio, columns=cols_orden)
```

```python
#Pasar los archivos a CSV como fue solicitado, les pongo el nombre "limpio" para no confundir con los transformados que son post aplicacion de one hot y normalizacion

X_train_limpio = pd.DataFrame(X_train_limpio)
X_train_limpio.to_csv(r'C:\Users\Mateio\Desktop\AA I\X_train_limpio.csv',index=False)

X_test_limpio = pd.DataFrame(X_test_limpio)
X_test_limpio.to_csv(r'C:\Users\Mateio\Desktop\AA I\X_test_limpio.csv',index=False)

y_train = pd.DataFrame(y_train)
y_train.to_csv(r'C:\Users\Mateio\Desktop\AA I\y_train.csv',index=False)

y_test = pd.DataFrame(y_test)
y_test.to_csv(r'C:\Users\Mateio\Desktop\AA I\y_test.csv',index=False)
```

## Segunda Parte. Transformar en Dummy Variables las nominales y normalización de las numéricas:

Nota: al transformar en dummy tenga cuidado de incluir todos las categorías posibles para cada variable categórica ... por suerte al inicio nos indicaron qué valores puede asumir cada variable ...  Con las numéricas hay menos problema.  

- Cuando cree el transformador de columnas ... tenga cuidado de no darle el mismo nombre que al del paso anterior ... no debería dar problemas, pero mejor ser precavido.

- A los DataFrames obtenidos después de transformar, llámelos: X_train_transformado, y_train, X_test_transformado, y_test
- **IMPORTANTE:** Al aplicar los transformadores, aplique primero el de las columnas numéricas y después el de las categóricas, éso facilitará la corrección.

```python
#Como fue indicado, primero realizo la normalizacion de las variables numericas
from sklearn.preprocessing import MinMaxScaler

t_norm=("normalizador",MinMaxScaler(feature_range=(0, 1)),cols_num) #Chequeado array cols_num
t_norm

transformador_norm = ColumnTransformer(transformers=[t_norm],remainder='passthrough')

transformador_norm.fit(X_train_limpio) #Entreno solo en el train

X_train_norm = transformador_norm.transform(X_train_limpio)  #Aplico la normalizacion a x train y x test

X_test_norm = transformador_norm.transform(X_test_limpio)


cols_orden=cols_num.tolist() + cols_cat.tolist()
X_train_norm=pd.DataFrame(X_train_norm, columns=cols_orden) 
X_test_norm=pd.DataFrame(X_test_norm, columns=cols_orden)
X_train_norm
```

```python
# Dummy para las Nominales: como conocemos cuáles son los valores posibles para cada variable, preferimos
# asignar manualmente los valores de cada categoria
from sklearn.preprocessing import OneHotEncoder

val_A1=['b','a']
val_A4=['u','y','l','t']
val_A5=['g','p','gg']
val_A6=['c','d','cc','i','j','k','m','r','q','w','x','e','aa','ff']
val_A7=['v','h','bb','j','n','z','dd','ff','o']
val_A9=['t','f']
val_A10=['t','f']
val_A12=['t','f']
val_A13=['g','p','s']

t_dummy=("onehot",OneHotEncoder(sparse=False,categories=[val_A1,val_A4,val_A5,val_A6,val_A7,val_A9,val_A10,val_A12,val_A13]),
         cols_cat) #Cols cat array chequeado que tiene el mismo orden que describi para asignar arriba

transformador_dummy= ColumnTransformer(transformers=[t_dummy],remainder='passthrough')

transformador_dummy.fit(X_train_norm) #Entreno sobre el train porque tenemos todos los valores posibles de cada categoria

X_train_transformado = transformador_dummy.transform(X_train_norm)

X_test_transformado = transformador_dummy.transform(X_test_norm)
```

### kNN

Para kNN, debemos seleccionar el mejor hiperparámetro "K", es decir, la cantidad de puntos vecinos que tenemos en cuenta en las cercanías para clasificar los grupos (que ya se conocen de antemano).

Para cada valor de K, voy a crear el modelo, entrenarlo y evaluarlo con F1 y Accuracy, tomando este último como el criterio para definir cual es el mejor K. Si el valor de Accuracy obtenido por el K es mejor que el anterior, lo voy a guardar en una variable para mostrar al final de las 50 iteraciones cual fue el mejor resultado y a que K corresponde.

#### Knn. Distance: uniform

```python
print("Evaluación modelo kNN en el Validation")

#Variables auxiliares donde se almacenarán los resultados del modelo elegido, las seteamos en 0 para empezar.
k_max=0
F1kNN_max=0
ACkNN_max=0

for k in k_valores:
    modelokNN = neighbors.KNeighborsClassifier(n_neighbors=k,weights='uniform') #Creo modelo
    modelokNN.fit(X_val_train, y_val_train)                                     #Entreno
    y_val_pred=modelokNN.predict(X_val_test)                                    #Pronostico
    ACkNN=accuracy_score(y_val_test, y_val_pred)                                #Obtengo AC
    F1kNN=f1_score(y_val_test,y_val_pred, average='macro')                      #Obtengo F1
    #print(str(k) + ' ' + str(F1kNN) + ' ' + str(ACkNN))
    if ACkNN>ACkNN_max:                 #Comparo Accuracy con el maximo y si es mayor, lo reemplazo.
        k_max=k                                                                 
        F1kNN_max=F1kNN
        ACkNN_max=ACkNN
        
print('Valor de k: ', k_max)                  #Hago print del mejor K y su respectivo Accuracy y F1.
print("Accuracy: ", round(ACkNN_max,6))
print("F1: ", round(F1kNN_max,6))
```

#### Knn. Distance: distance

```python
print("Evaluación modelo kNN en el Validation")

#Variables auxiliares donde se almacenarán los resultados del modelo elegido, las seteamos en 0 para empezar.
k=0
k_maxd=0
F1kNNd_max=0
ACkNNd_max=0

for k in k_valores:        
    modelokNN = neighbors.KNeighborsClassifier(n_neighbors=k,weights='distance') #Creo modelo
    modelokNN.fit(X_val_train, y_val_train)                                      #Entreno
    y_val_pred=modelokNN.predict(X_val_test)                                     #Pronostico
    ACkNNd=accuracy_score(y_val_test, y_val_pred)                                #Obtengo AC
    F1kNNd=f1_score(y_val_test,y_val_pred, average='macro')                      #Obtengo F1
    #print(str(k) + ' ' + str(F1kNNd) + ' ' + str(ACkNNd))
    if ACkNNd>ACkNNd_max:                 #Comparo Accuracy con el maximo y si es mayor, lo reemplazo.
        k_maxd=k                                                                 
        F1kNNd_max=F1kNNd
        ACkNNd_max=ACkNNd
        
print('Valor de k: ', k_maxd)               #Hago print del mejor K y su respectivo Accuracy y F1.
print("Accuracy: ", round(ACkNNd_max,6))
print("F1: ", round(F1kNNd_max,6))
```

### Arbol de decisión

Los arboles de decision, tienen el hiperparámetro "max-depth", que significa la profunidad del arbol. Siguiendo un criterio similar al anterior, vamos a iterar para obtener la profundidad que nos provea el mejor AC, sin caer en overfitting.

```python
#Igual que en el modelo anterior, creo array de numeros del 1 al 100 para luego iterar.
profundidades=np.arange(1,101,1)
```

```python
#Creo variables auxiliares donde se almacenarán los resultados del modelo elegido, las seteamos en 0 para empezar.

p_max=0               #Aca guardaremos la profundidad del arbol que mejor resuelve el problema (mayor AC)
ACT_max=0
F1T_max=0

print("Evaluación modelo Árbol de Decisión en el Validation")

for p in profundidades: 
    modeloT=DecisionTreeClassifier(max_depth=p, random_state=123)        #Creamos el modelo con profundidad p
    modeloT.fit(X_val_train, y_val_train)                                # Lo entrenamos
    y_val_pred = modeloT.predict(X_val_test)                             # Pronosticamos los valores de y para los valores del X_test
    ACT=accuracy_score(y_val_test, y_val_pred)
    F1T=f1_score(y_val_test,y_val_pred, average='macro')
    #print(str(p) + ' ' + str(F1T) + ' ' + str(ACT))
    if ACT>ACT_max:                                                      #Utilizo accuracy como el parámetro que decide.
        p_max=p
        F1T_max=F1T
        ACT_max=ACT
    
print('Valor de profundidad ', p_max)
print("Accuracy: ", round(ACT_max,6))
print("F1: ", round(F1T_max,6))
```
